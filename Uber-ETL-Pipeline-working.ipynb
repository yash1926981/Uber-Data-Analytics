```python
payment_type_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}

payment_type_dim = df[['payment_type']].reset_index(drop=True)
payment_type_dim['payment_type_id'] = payment_type_dim.index
payment_type_dim['payment_type_name']=payment_type_dim['payment_type'].map(payment_type_name)
payment_type_dim=payment_type_dim[['payment_type_id','payment_type','payment_type_name']]
```

```python
#Factable creation

fact_table = df.merge(passenger_count_dim,left_on='trip_id',right_on='passenger_count_id')\
               .merge(trip_distance_dim,left_on='trip_id',right_on='trip_distance_id')\
               .merge(rate_code_dim,left_on='trip_id',right_on='rate_code_id')\
               .merge(pickup_location_dim,left_on='trip_id',right_on='pickup_location_id')\
               .merge(dropoff_location_dim,left_on='trip_id',right_on='dropoff_location_id')\
               .merge(datetime_dim,left_on='trip_id',right_on='datetime_id')\
               .merge(payment_type_dim,left_on='trip_id',right_on='payment_type_id')\
               [['trip_id','VendorID','datetime_id','passenger_count_id',
                 'trip_distance_id','rate_code_id','store_and_fwd_flag','pickup_location_id','dropoff_location_id',
                 'payment_type_id','fare_amount','extra','mta_tax','tip_amount','tolls_amount',
                 'improvement_surcharge','total_amount']]
```

```python
# Write in transformer block
passenger_count_dim['type'] = 'passenger_count'
trip_distance_dim['type'] = 'trip_distance'
rate_code_dim['type']='rate_code'
payment_type_dim['type'] = 'payment_type'
pickup_location_dim['type'] = 'pickup'
dropoff_location_dim['type']='dropoff'
datetime_dim['type']  = 'datetype'
fact_table['type'] = 'fact'


combined_df = pd.concat([passenger_count_dim,trip_distance_dim,rate_code_dim,payment_type_dim,datetime_dim,pickup_location_dim,dropoff_location_dim,fact_table], ignore_index=True,sort=False)
```

```python
#Splitting of dataframes
fact_table = combined_df[combined_df['type'] == 'fact'].drop(columns=['type']).dropna(axis=1, how='all')
passenger_count_dim = combined_df[combined_df['type'] == 'passenger_count'].drop(columns=['type']).dropna(axis=1, how='all')
trip_distance_dim = combined_df[combined_df['type'] == 'trip_distance'].drop(columns=['type']).dropna(axis=1, how='all')
rate_code_dim = combined_df[combined_df['type'] == 'rate_code'].drop(columns=['type']).dropna(axis=1, how='all')
payment_type_dim = combined_df[combined_df['type'] == 'payment_type'].drop(columns=['type']).dropna(axis=1, how='all')
pickup_location_dim = combined_df[combined_df['type'] == 'pickup'].drop(columns=['type']).dropna(axis=1, how='all')
dropoff_location_dim = combined_df[combined_df['type'] == 'dropoff'].drop(columns=['type']).dropna(axis=1, how='all')
datetime_dim = combined_df[combined_df['type'] == 'datetype'].drop(columns=['type']).dropna(axis=1, how='all')
```

```python
#Dictionaries

fact = {
'trip_id':int,  
'VendorID':int,  
'datetime_id':int, 
'passenger_count_id':int, 
'trip_distance_id':int,
'rate_code_id':int,   
'pickup_location_id':int,  
'dropoff_location_id':int,  
 'payment_type_id':int,
'fare_amount':float,
'extra':float,
'mta_tax':float,
'tip_amount':float,
'tolls_amount':float,
'improvement_surcharge':float,
'total_amount':float
}

datetype = {
'datetime_id':int,         
'pick_hour':int,        
'pick_day':int,     
'pick_month':int,         
 'pick_year':int,         
 'pick_weekday':int,         
 'drop_hour':int,         
 'drop_day':int,         
 'drop_month':int,         
 'drop_year':int,         
 'drop_weekday':int
}

passenger_count = {

 'passenger_count_id':int,
'passenger_count':int

}

trip_distance = {
 'trip_distance_id':int,
'trip_distance':float
    
}

rate_code = {
'rate_code_id':int, 
'RatecodeID':int
}

payment_type = {
'payment_type_id':int,
'payment_type':int

}

pickup = {
'pickup_location_id':int,  
'pickup_latitude':float,
'pickup_longitude':float

}

dropoff = {
'dropoff_location_id':int,  
'dropoff_latitude':float,
'dropoff_longitude':float

}
```

```python
#Datatype conversion
fact_table = fact_table.astype(fact)
datetime_dim = datetime_dim.astype(datetype)
passenger_count_dim = passenger_count_dim.astype(passenger_count)
trip_distance_dim = trip_distance_dim.astype(trip_distance)
rate_code_dim = rate_code_dim.astype(rate_code)
payment_type_dim = payment_type_dim.astype(payment_type)
pickup_location_dim = pickup_location_dim.astype(pickup)
dropoff_location_dim = dropoff_location_dim.astype(dropoff)
```

```python
#Columns order list
fact_table_columns_order = ['trip_id','VendorID','datetime_id','passenger_count_id',
                 'trip_distance_id','rate_code_id','store_and_fwd_flag','pickup_location_id','dropoff_location_id',
                 'payment_type_id','fare_amount','extra','mta_tax','tip_amount','tolls_amount',
                 'improvement_surcharge','total_amount']

datetime_dim_columns_order = ['datetime_id','tpep_pickup_datetime','pick_hour','pick_day','pick_month','pick_year','pick_weekday',
                             'tpep_dropoff_datetime','drop_hour','drop_day','drop_month','drop_year','drop_weekday']

passenger_dim_columns_order = ['passenger_count_id','passenger_count']

trip_distance_dim_columns_order = ['trip_distance_id','trip_distance']

rate_code_dim_columns_order = ['rate_code_id','RatecodeID','rate_code_name']

payment_type_dim_columns_order = ['payment_type_id','payment_type','payment_type_name']

pickup_location_dim_columns_order = ['pickup_location_id','pickup_latitude','pickup_longitude']

dropoff_location_dim_columns_order = ['dropoff_location_id','dropoff_latitude','dropoff_longitude']


```

```python
#Setting up the order
fact_table = fact_table[fact_table_columns_order]
datetime_dim = datetime_dim[datetime_dim_columns_order]
passenger_count_dim = passenger_count_dim[passenger_dim_columns_order]
trip_distance_dim = trip_distance_dim[trip_distance_dim_columns_order]
rate_code_dim = rate_code_dim[rate_code_dim_columns_order]
payment_type_dim = payment_type_dim[payment_type_dim_columns_order]
pickup_location_dim = pickup_location_dim[pickup_location_dim_columns_order]
dropoff_location_dim = dropoff_location_dim[dropoff_location_dim_columns_order]
```

```python
#Resetting index
fact_table = fact_table.reset_index(drop=True)
datetime_dim = datetime_dim.reset_index(drop=True)
passenger_count_dim = passenger_count_dim.reset_index(drop=True)
trip_distance_dim = trip_distance_dim.reset_index(drop=True)
rate_code_dim = rate_code_dim.reset_index(drop=True)
payment_type_dim = payment_type_dim.reset_index(drop=True)
pickup_location_dim = pickup_location_dim.reset_index(drop=True)
dropoff_location_dim = dropoff_location_dim.reset_index(drop=True)
```

```python
#Setting index
fact_table['trip_id'] = fact_table.index
datetime_dim['datetime_id'] = datetime_dim.index
passenger_count_dim['passenger_count_id'] = passenger_count_dim.index
trip_distance_dim['trip_distance_id'] = trip_distance_dim.index
rate_code_dim['rate_code_id'] = rate_code_dim.index
payment_type_dim['payment_type_id'] = payment_type_dim.index
pickup_location_dim['pickup_location_id'] = pickup_location_dim.index
dropoff_location_dim['dropoff_location_id'] = dropoff_location_dim.index
```
